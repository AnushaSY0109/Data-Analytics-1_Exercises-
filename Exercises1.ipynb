```python
#Descriptive statistical analysis on dataset
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
file_path='/content/sales_data.csv'
df=pd.read_csv(file_path)
mean=np.mean(df['price'])
print("Mean_of_price=",mean)
Median=np.median(df['price'])
Mode = df['price'].mode()[0]
print("Median_of_price=",Median)
print("Mode_of_price=",Mode)
Var=np.var(df['price'])
print("Variance=",Var)
SD=np.std(df['price'])
print("Standard Deviation=",SD)
range= np.ptp(df['price'])
print("Range=",range)
```

```python
#plotting the histogram
plt.figure(figsize=(10,6))
plt.hist(df['price'],bins=20,edgecolor='red',color='green')
plt.suptitle('Histograms of Sales_Price',fontsize=18)
plt.tight_layout(rect=[0,0,1,0.95])
plt.show()
```
